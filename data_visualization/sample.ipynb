```python
def plot_xyz(sample, title):
    fig, ax = plt.subplots(6, 1, figsize=(30, 17))

    enmo_non_zero = sample.filter(pl.col("enmo")!=0)
    ax[0].plot(enmo_non_zero["step"], enmo_non_zero["X"], color="orange")
    enmo_zero = sample.filter(pl.col("enmo")==0)
    ax[0].plot(enmo_zero["step"], enmo_zero["X"], color="blue")
    ax[0].set_title("X")
    
    # ax[1].plot(sample["Y"])
    # ax[1].set_title("Y")
    
    # ax[2].plot(sample["Z"])
    # ax[2].set_title("Z")

    # ax[3].plot(sample["enmo"])
    # ax[3].set_title("enmo")
    
    # ax[4].plot(sample["light"])
    # ax[4].set_title("light")
    
    # ax[5].plot(sample["non-wear_flag"])
    # ax[5].set_title("non-wear_flag")

    fig.suptitle(title)
    plt.show()
```

```python
plot_xyz(sample=sample, title="sample")
```
